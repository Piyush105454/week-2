```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes

# Load the weather data from the CSV file
weather_data = pd.read_csv('weather_data_extended.csv')

# Create a WindroseAxes object
ax = WindroseAxes.from_ax()

# Plot the wind rose
ax.bar(weather_data['Wind Speed (kph)'], weather_data['Wind Speed (kph)'], normed=True, opening=0.8, edgecolor='white')

# Set title and legend
ax.set_title('Wind Rose for Metropolis')
ax.legend(title='Wind Speed (kph)')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual dataset)
data = {
    'Location': ['New York', 'London', 'Paris', 'Tokyo', 'Hong Kong', 'Dubai', 'Rome', 'Sydney'],
    'Sunny Days': [15, 10, 12, 20, 25, 30, 18, 22],
    'Partly Cloudy Days': [10, 12, 14, 8, 5, 3, 11, 9],
    'Overcast Days': [5, 8, 6, 2, 1, 0, 4, 3],
    'Precipitation Days': [8, 10, 9, 7, 5, 3, 6, 8]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Plot sunny, partly cloudy, overcast, and precipitation days for all locations
barWidth = 0.2
locations = range(len(df['Location']))

plt.bar(locations, df['Sunny Days'], color='yellow', width=barWidth, label='Sunny Days')
plt.bar([i + barWidth for i in locations], df['Partly Cloudy Days'], color='lightblue', width=barWidth, label='Partly Cloudy Days')
plt.bar([i + 2 * barWidth for i in locations], df['Overcast Days'], color='gray', width=barWidth, label='Overcast Days')
plt.bar([i + 3 * barWidth for i in locations], df['Precipitation Days'], color='blue', width=barWidth, label='Precipitation Days')

# Add labels and title
plt.xlabel('Location')
plt.ylabel('Number of Days')
plt.title('Monthly Weather Distribution for Different Cities')
plt.xticks([i + 1.5 * barWidth for i in locations], df['Location'])
plt.legend()

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the weather data from the CSV file
df = pd.read_csv('weather_data_extended.csv')

# Define the list of locations
locations = ['New York', 'London', 'Paris', 'Tokyo', 'Hong Kong', 'Dubai', 'Rome', 'Sydney']

# Plotting the weather data for all specified locations
plt.figure(figsize=(12, 8))

# Define the parameters to plot
parameters = ['Temperature (Â°C)', 'Humidity (%)', 'Wind Speed (kph)', 'Cloud Cover (%)', 'Pressure (mb)']

# Plot each parameter for all specified locations
for location in locations:
    location_data = df[df['Location'] == location]
    for parameter in parameters:
        plt.plot(location_data['Observation Time'], location_data[parameter], marker='o', label=f'{location} - {parameter}')

# Add labels and title
plt.xlabel('Observation Time')
plt.ylabel('Value')
plt.title('Weather Data for All Specified Locations')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
