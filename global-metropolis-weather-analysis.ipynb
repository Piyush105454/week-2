# Part 3.1: Wind Rose Analysis

**Wind Rose Description**

A wind rose is a graphical tool used to visualize the frequency and distribution of wind direction and speed at a particular location over a specific period. It is typically presented in a circular format, where the direction of the wind is plotted radially around the circle, and the frequency or speed of the wind is represented by the length or color of the bars.

**Components of a Wind Rose:**

*Directions:* The compass directions (e.g., North, South, East, West) are divided into sectors around the circle, with each sector representing a range of wind directions.

*Frequency/Speed Bars:* Bars extending from the center of the circle outward represent the frequency or speed of the wind from each direction. The length or color intensity of the bars indicates the magnitude of the wind.

**Interpretation:**

Wind roses provide valuable insights into prevailing wind patterns and variability at a specific location.
They help in understanding wind behavior, which is crucial for various applications such as climate studies, urban planning, agriculture, and renewable energy projects.
By analyzing wind roses, one can identify dominant wind directions, seasonal variations, and potential wind energy resources in a given area.
Usage:

**Wind energy assessment:** Wind roses aid in selecting suitable locations for wind turbine installations by assessing wind speed and direction distributions.
Navigation and maritime operations: They assist sailors and navigators in planning routes and avoiding adverse weather conditions.

**Environmental studies:** Wind roses help researchers analyze air pollution dispersion patterns and atmospheric circulation.
Aviation and meteorology: They provide information for flight planning, airport operations, and weather forecasting.
In summary, wind roses serve as valuable tools for understanding and visualizing wind patterns, which have diverse applications across various industries and disciplines.

```python
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes

# Load the weather data from the CSV file
weather_data = pd.read_csv('/kaggle/input/global-metropolis-weather-dataset/weather_data_extended.csv')

# Create a WindroseAxes object
ax = WindroseAxes.from_ax()

# Plot the wind rose
ax.bar(weather_data['Wind Speed (kph)'], weather_data['Wind Speed (kph)'], normed=True, opening=0.8, edgecolor='white')

# Set title and legend
ax.set_title('Wind Rose for Metropolis')
ax.legend(title='Wind Speed (kph)')

# Show the plot
plt.show()
```

```python
def plot_wind_rose(city_data, city_name):
    ax = WindroseAxes.from_ax()
    ax.bar(city_data['Wind Speed (kph)'], city_data['Wind Speed (kph)'], normed=True, opening=0.8, edgecolor='white')
    ax.set_title(f'Wind Rose - {city_name}')
    ax.legend(title='Wind Speed (kph)')
    plt.show()

# Iterate over each city and plot wind rose
for city_name, city_data in df.groupby('Location'):
    plot_wind_rose(city_data, city_name)
```

# Part 3.2: Average Temperatures 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual dataset)
data = {
    'Location': ['New York', 'London', 'Paris', 'Tokyo', 'Hong Kong', 'Dubai', 'Rome', 'Sydney'],
    'Mean Daily Max (°C)': [20, 18, 22, 25, 28, 30, 23, 22],
    'Mean Daily Min (°C)': [10, 8, 12, 15, 18, 20, 13, 12],
    'Hot Days (°C)': [30, 28, 32, 35, 38, 40, 33, 32],
    'Cold Nights (°C)': [5, 3, 7, 10, 12, 15, 8, 7]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Plot mean daily maximum and minimum temperatures
plt.bar(df['Location'], df['Mean Daily Max (°C)'], color='red', label='Mean Daily Max')
plt.bar(df['Location'], df['Mean Daily Min (°C)'], color='blue', label='Mean Daily Min')

# Plot hot days and cold nights
plt.plot(df['Location'], df['Hot Days (°C)'], linestyle='--', color='red', label='Hot Days')
plt.plot(df['Location'], df['Cold Nights (°C)'], linestyle='--', color='limegreen', label='Cold Nights')

# Add labels and title
plt.xlabel('Location')
plt.ylabel('Temperature (°C)')
plt.title('Average Temperatures for Different Cities')
plt.legend()

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Part 3.3: Monthly Weather Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace this with your actual dataset)
data = {
    'Location': ['New York', 'London', 'Paris', 'Tokyo', 'Hong Kong', 'Dubai', 'Rome', 'Sydney'],
    'Sunny Days': [15, 10, 12, 20, 25, 30, 18, 22],
    'Partly Cloudy Days': [10, 12, 14, 8, 5, 3, 11, 9],
    'Overcast Days': [5, 8, 6, 2, 1, 0, 4, 3],
    'Precipitation Days': [8, 10, 9, 7, 5, 3, 6, 8]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Plot sunny, partly cloudy, overcast, and precipitation days for all locations
barWidth = 0.2
locations = range(len(df['Location']))

plt.bar(locations, df['Sunny Days'], color='yellow', width=barWidth, label='Sunny Days')
plt.bar([i + barWidth for i in locations], df['Partly Cloudy Days'], color='lightblue', width=barWidth, label='Partly Cloudy Days')
plt.bar([i + 2 * barWidth for i in locations], df['Overcast Days'], color='gray', width=barWidth, label='Overcast Days')
plt.bar([i + 3 * barWidth for i in locations], df['Precipitation Days'], color='blue', width=barWidth, label='Precipitation Days')

# Add labels and title
plt.xlabel('Location')
plt.ylabel('Number of Days')
plt.title('Monthly Weather Distribution for Different Cities')
plt.xticks([i + 1.5 * barWidth for i in locations], df['Location'])
plt.legend()

# Show plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Part 3.4: Weather Data for All Locations

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the weather data from the CSV file
df = pd.read_csv('/kaggle/input/global-metropolis-weather-dataset/weather_data_extended.csv')

# Define the list of locations
locations = ['New York', 'London', 'Paris', 'Tokyo', 'Hong Kong', 'Dubai', 'Rome', 'Sydney']

# Plotting the weather data for all specified locations
plt.figure(figsize=(12, 8))

# Define the parameters to plot
parameters = ['Temperature (°C)', 'Humidity (%)', 'Wind Speed (kph)', 'Cloud Cover (%)', 'Pressure (mb)']

# Plot each parameter for all specified locations
for location in locations:
    location_data = df[df['Location'] == location]
    for parameter in parameters:
        plt.plot(location_data['Observation Time'], location_data[parameter], marker='o', label=f'{location} - {parameter}')

# Add labels and title
plt.xlabel('Observation Time')
plt.ylabel('Value')
plt.title('Weather Data for All Specified Locations')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

# Part 3.5: Maximum Temperatures

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the weather data from the CSV file
weather_data = pd.read_csv('/kaggle/input/global-metropolis-weather-dataset/weather_data_extended.csv')

# Convert the 'Observation Time' column to datetime format
weather_data['Observation Time'] = pd.to_datetime(weather_data['Observation Time'])

# Extract month from the 'Observation Time' column
weather_data['Month'] = weather_data['Observation Time'].dt.month

# Group the data by month and calculate the maximum temperature for each month
max_temp_per_month = weather_data.groupby(['Month', 'Location'])['Temperature (°C)'].max().reset_index()

# Determine the city with the highest maximum temperature
city_with_highest_temp = max_temp_per_month.loc[max_temp_per_month['Temperature (°C)'].idxmax(), 'Location']

# Define temperature ranges
temperature_ranges = [0, 10, 20, 30, 40, 50]

# Count the number of days per month that reach certain temperatures
temperature_counts = pd.cut(max_temp_per_month['Temperature (°C)'], bins=temperature_ranges, include_lowest=True).value_counts().sort_index()

# Plot the maximum temperature distribution
plt.figure(figsize=(10, 6))
temperature_counts.plot(kind='bar', color='skyblue')
plt.title('Maximum Temperature Distribution Across Temperature Ranges')
plt.xlabel('Temperature Range (°C)')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

print(f"The city with the highest maximum temperature is: {city_with_highest_temp}")
```
